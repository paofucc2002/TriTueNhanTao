```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
url = 'mtcars.csv'
```

```python
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.hist()
```

```python
plt.show()
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
```

```python
print('Giá trị intercept:',model.intercept_)
print('Giá trị coeffiant:0', model.coef_)
```

```python
print('Độ chính xác:',model.score(X=pd.DataFrame(data['wt']),y=data['mpg']))
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2]).T
model.fit(X,y=data['mpg'])
print('Intercept là:',model.intercept_)
print('Coeffiant là:',model.coef_)
print('Độ chính xác là:',model.score(X,y=data['mpg']))
```

```python
range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3]).T
model.fit(X,y=data['mpg'])
print('Intercept là:',model.intercept_)
print('Coeffiant là:',model.coef_)
print('Độ chính xác là:',model.score(X,y=data['mpg']))
```

```python
range = np.arange(1.5,5.5,0.1)
X_range=pd.DataFrame([range,range**2,range**3]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
```
