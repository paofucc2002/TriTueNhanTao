```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
url = '50_Startups.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.hist()
plt.show()
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Profit']),y=data['R&D Spend'])
```

```python
print('Giá trị intercept:',model.intercept_)
print('Giá trị coeffiant:0', model.coef_)
print('Độ chính xác:',model.score(X=pd.DataFrame(data['Profit']),y=data['R&D Spend']))
```

```python
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['Profit']))
plt.plot(data['Profit'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['Profit'],data['Profit']**2]).T
model.fit(X,y=data['R&D Spend'])
print('Intercept là:',model.intercept_)
print('Coeffiant là:',model.coef_)
print('Độ chính xác là:',model.score(X,y=data['R&D Spend']))
```

```python
range = np.arange(1000,210000,5000)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['Profit'],data['Profit']**2,data['Profit']**3,data['Profit']**4,data['Profit']**5]).T
model.fit(X,y=data['R&D Spend'])
print('Intercept là:',model.intercept_)
print('Coeffiant là:',model.coef_)
print('Độ chính xác là:',model.score(X,y=data['R&D Spend']))
```

```python
range = np.arange(1000,210000,5000)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```
